```python
src_path = 'D:/Python Coding Ninjas/Text Classification Project/20_newsgroups_train/'
```

```python
d = {}
```

```python
from sklearn.feature_extraction.stop_words import ENGLISH_STOP_WORDS
stopwords = ENGLISH_STOP_WORDS
```

```python
newsgroups = os.listdir()
```

```python
# making the dictionary
for n_g in newsgroups:
    d[n_g] = {}
    new_path = os.path.join(src_path,n_g)
    for roots,dirs,files in os.walk(new_path):
        for file in files:
            read_f = open(os.path.join(new_path,file),'r')
            line = read_f.read()
            for word in line.split():
                if word in stopwords:
                    continue
                d[n_g][word] = {}
                d[n_g][word] = 0
```

```python
# inserting values in dictionary
for n_g in newsgroups:
    new_path = os.path.join(src_path,n_g)
    for roots,dirs,files in os.walk(new_path):
        for file in files:
            read_f = open(os.path.join(new_path,file),'r')
            line = read_f.read()
            for word in line.split():
                if word in stopwords:
                    continue
                d[n_g][word] += 1
```

```python
keys = [i for i in d.keys()]
```

```python
#make a copy of the dictionary
dict1 = d
```

```python
dict2 = {}
keys = dict1.keys()
for key in keys:
    for i in dict1[key]:
        dict2[i] = 0
```

```python
keys = dict1.keys()
for key in keys:
    for i in dict1[key]:
        dict2[i] += dict1[key][i]
```

```python
sorted_dict2 = {r: dict2[r] for r in sorted(dict2, key=dict2.get, reverse=True)}
```

```python
words = list(sorted_dict2.keys())
```
